```python
import samna
print(f"samna version: {samna.__version__}")
import time

device = samna.device.open_device('XyloAudio3TestBoard')

# get device configuration
ioc = device.get_io_control_module()
io = device.get_io_module()
model = device.get_model()

# power off
ioc.write_config(0, 0x0)
time.sleep(1)

# main clock disable
io.write_config(0x0008, 0)
time.sleep(1)

# power on
ioc.write_config(0, 0x1B)
time.sleep(1)

# main clock enable
io.write_config(0x0008, 1)
time.sleep(1)

# Set a board switch to route microphone PDM data rather than FPGA PDM data to Xylo by calling io.write_config(0x12, 1)
io.write_config(0x0012, 0)
time.sleep(1)

# Enable real-time mode by calling io.write_config(0x31, 2)
io.write_config(0x31, 2)

# PDM clock enable
io.write_config(0x0029, 1)

time.sleep(1)


# create access to xylo input/output
source = samna.graph.source_to(model.get_sink_node())
sink   = samna.graph.sink_from(model.get_source_node())


# write registers for real-time mode -- values copied from Xin's tests
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["ctrl1"], data=0x143_011C)])
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["ctrl2"], data=0x9_0902)])
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["ctrl3"], data=0x901)])

source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["tr_wrap"], data=500000)])
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["hm_tr_wrap"], data=0x0)])

source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["clk_ctrl"], data=0x011)])
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["clk_div"], data=0x18_0000)])

source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["pwr_ctrl1"], data=0x33_00FF)])
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["pwr_ctrl2"], data=0x3_0001)])
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["pwr_ctrl3"], data=0xFFFF_FFFF)])
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["pwr_ctrl4"], data=0xFFFF_0001)])

source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["pad_ctrl"], data=0x00)])
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["ie1"], data=0x0)])
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["ie2"], data=0x1)])
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["out_ctrl"], data=0x70)])

source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["monsel"], data=0x0)])
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["mon_grp_sel"], data=0x1F1F_0773)])
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["dbg_ctrl1"], data=0x0)])
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["dbg_stat1"], data=0x0)])
source.write([samna.xyloAudio3.event.WriteRegisterValue(address=registers["dfe_ctrl"], data=0x110)])

print("configured!")
```

```python
def read_register(addr):
    source.write([samna.xyloAudio3.event.ReadRegisterValue(address = addr)])
    events = sink.get_n_events(1, 3000)
    assert(len(events) == 1)
    return events[0].data
    
print("\n# ======= read all registers =======\n")

for register in registers:
    print(f"{register}       : 0x{format(read_register(registers[register]),'_X')}")
print()
```

```python
print("\n# ======= read non-zero registers =======\n")

for register in registers:
    if (read_register(registers[register]) != 0):
        print(f"{register}       : 0x{format(read_register(registers[register]),'_X')}")
print()
```
